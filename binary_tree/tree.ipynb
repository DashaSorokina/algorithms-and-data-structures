InOrder Traversal

```python
def in_order(root):
    if root is None:
        return
    in_order(root.left)
    print(root.value, end=' ')
    in_order(root.right)
```

```python
in_order(root)
```

Preorder Traversal

```python
def pre_order(root):
    if root is None:
        return
    else:
        print(root.value, end=' ')
        pre_order(root.left)
        pre_order(root.right)
```

```python
pre_order(root)
```

Postorder Traversal

```python
def post_order(root):
    if root is None:
        return
    else:
        post_order(root.right)
        post_order(root.left)
        print(root.value, end=' ')
      
```

```python
post_order(root)
```

BFS

```python
def bfs(root):
    if root is None:
        return 

    visited = []
    q = deque()
    q.append(root)

    while q:
        node = q.popleft()
        visited.append(node.value)

        if node.left:
            q.append(node.left)

        if node.right:
            q.append(node.right)

    return visited
```

```python
bfs(root)
```
